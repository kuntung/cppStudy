# 三只小猪称重

通过用户输入的三只小猪的体重，对比输出三只小猪的体重排序

```c++
#include<iostream>

using namespace std;
```

```c++
//初始化三只小猪的体重
float pigA = 0.0f;
float pigB = 0.0f;
float pigC = 0.0f;
```

```c++
cout<<"请分别输入三只小猪的体重：";
cin>>pigA;
cin>>pigB;
cin>>pigC;
```

```c++
//通过多重if结构判断三只小猪的体重，并输出体重排序
if(pigA >= pigB && pigA >= pigC){
        cout<<"这三只小猪中最重的是A："<<pigA<<" kg.";
    if(pigB >= pigC){
        cout<<"最轻的是小猪C："<<pigC<<" kg。";
    }
    else{
        cout<<"最轻的是小猪B："<<pigB<<" kg。";
    }
}
else 
    if(pigB >= pigA && pigB >= pigC){
        cout<<"这三只小猪中最重的是B："<<pigB<<" kg.";
}
    else{
        cout<<"这三只小猪中最重的是C："<<pigC<<" kg.";
}
```
