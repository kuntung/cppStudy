# 点和圆的关系
设计一个圆类和点类，计算点和圆的关系

```c++
#include<iostream>
#include<string>

using namespace std;

const float PI = 3.14;
```

```c++
//设计点类
class Point{
    public:
        void setX(float x){
            m_x = x;
        }
        void setY(float y){
            m_y = y;
        }
        float getX(){
            return m_x;
        }
        float getY(){
            return m_y;
        }
    private:
        float m_x;
        float m_y;
};
```

```c++
//设计圆类
class Circle{
    //将成员函数设置为public，以用来访问和调用私有化的成员变量
    public:
        void setR(float radiu){
            m_R = radiu;
        }
        void setCenter(Point center){
            m_Center = center;
        }
        float getR(){
            return m_R;
        }
        Point getCenter(){
            return m_Center;
        }
        void showLocation(Point p1){
            float dis = pow((p1.getX() - getCenter().getX()),2) + pow((p1.getY() - getCenter().getY()),2);
            float squareR = pow(getR(), 2);
            if(dis == 0){
                cout<<"点在圆心"<<"\n";
            }
            else if(dis < squareR){
                cout<<"点在圆内"<<"\n";
            }
            else if(squareR == dis){
                cout<<"点在圆上"<<"\n";
            }
            else{
                cout<<"点在圆外"<<"\n";
            }
        }
            
    //先将成员属性设置为私有
    private:
        float m_R;
        Point m_Center;//在类中，可以让另一个类作为
};
```

```c++
Point p1;
p1.setX(3.2);
p1.setY(6.4);

Point p2;
p2.setX(3.2);
p2.setY(6.4);
Circle c1;
c1.setR(3);
c1.setCenter(p1);
c1.showLocation(p2);
// c1.getCenter().getX()
```
