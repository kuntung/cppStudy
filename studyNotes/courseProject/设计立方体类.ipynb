# 设计立方体类
要求：
1. 求出立方体的面积和体积
2. 分别用全局函数和成员函数判断两个立方体是否相等

```c++
#include<iostream>
#include<string>

using namespace std;
```

```c++
class Cube{
    public://提供一些公用借口
        float m_Area(){
            return 2 * m_L * m_H + 2 * m_L * m_W + 2 * m_H * m_W;
        }
        float m_Volume(){
            return m_L * m_H * m_W;
        }
        void setL(float length){
            m_L = length;
        }
        void setH(float height){
            m_H = height;
        }
        void setW(float width){
            m_W = width;
        }
        float getL(){
            return m_L;
        }
        float getW(){
            return m_W;
        }
        float getH(){
            return m_H;
        }
        bool isSameByClass(Cube &c2){
            if(m_L == c2.getL() && m_W == c2.getW() && m_H == c2.getH()){
                return true;}
            else{
                return false;
                }
        }
    private://将成员属性设置为私有
        float m_H;
        float m_L;
        float m_W;
};
```

```c++
//全局函数判断两个立方体是否相等
bool isSame(Cube &c1, Cube &c2）{//传入两个立方体的引用

    if(c1.getH() == c2.getH() && c1.getW() == c2.getW() && c1.getL() == c2.getL()){
        return true;
    }
    else{
        return false;
    }

}
```

```c++
Cube c1;
c1.setW(23);
c1.setL(13.2);
c1.setH(3.4);
float area = c1.m_Area();

Cube c2;
c2.setW(23);
c2.setL(13.2);
c2.setH(3.4);
bool sameFLAG = c1.isSameByClass(c2);
if(sameFLAG){
    cout<<"这两个立方体相同"<<"\n";
}

isSame(c1, c2);
```
