```c++
//vector
#include<iostream>
#include<vector>
#include<algorithm>

using namespace std;
```

```c++
//vector
class Person{
    public:
        Person(string name, int age){
            this->m_name = name;
            this->m_age = age;
        }
        
        void show_info(){
            cout<<":"<<this->m_name<<"\n";
        }
        string m_name;
        int m_age;
};
```

```c++
void test04(){
    vector<Person> v;
    
    Person p1("A", 10);
    Person p2("B", 20);
    Person p3("C", 30);
    Person p4("D", 40);    
    Person p5("E", 50);
    
    //
    v.push_back(p1);
    v.push_back(p2);
    v.push_back(p3);
    v.push_back(p4);
    v.push_back(p5);
    
    //
    for(vector<Person>::iterator it = v.begin(); it != v.end(); it++){
        cout<<""<<(*it).m_name<<"\n";
        cout<<""<<it->m_age<<endl;
    }
}

test04();
```

```c++
//
void test05(){
    vector<Person*> v;
    
    Person p1("A", 10);
    Person p2("B", 20);
    Person p3("C", 30);
    Person p4("D", 40);    
    Person p5("E", 50);
    
    //
    v.push_back(&p1);
    v.push_back(&p2);
    v.push_back(&p3);
    v.push_back(&p4);
    v.push_back(&p5);
    
    //
    //*it
    for(vector<Person*>::iterator it = v.begin(); it != v.end(); it++){
        cout<<""<<(*it)->m_name<<"\n";
        cout<<""<<(*it)->m_age<<endl;
    }
}

test05();
```

```c++
//vector
#include<iostream>
#include<vector>
#include<algorithm>

using namespace std;
```

```c++
//
void test06(){
    vector<vector<int>> v;
    
    //
    vector<int> v1;
     vector<int> v2;
    
    //vector
    v1.push_back(10);
    v1.push_back(20);
    
    v2.push_back(30);
    v2.push_back(40);
    //vectorvector
    v.push_back(v1);
    v.push_back(v2);
    for(vector<vector<int>>::iterator it = v.begin(); it != v.end(); it++){
        //*itvector<T>T
        cout<<""<<endl;
        for(vector<int>::iterator sub_it = (*it).begin(); sub_it !=(*it).end(); sub_it++){
                cout<<*sub_it<<endl;
                }
         }

}

test06();
```

### string
`string`
1. `=`
2. `assign`




```c++
//string
void test02(){
    string str1("hello world");
    cout<<"str1 = "<<str1<<endl;
    //string& operator = (const char * s)
    str1 = "";
    cout<<"str1 = "<<str1<<endl;
    
    string str2;
    //string& operator = (const char &s)
    str2 = str1;
    cout<<"str2 = "<<str2<<endl;
    
    string str3;
    //string& operator =( char c)

    str3 = 'a';
    cout<<"str3 = "<<str3<<endl;
    
    //`assign()`
    string str4;
    str4.assign("");
    cout<<"str4 = "<<str4<<endl;
    
    
     
}

test02();
```
