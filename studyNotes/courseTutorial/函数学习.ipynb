# 函数
>作用：封装代码，减少重复代码

## 函数的定义
1. 返回值类型
2. 函数名：函数调用时使用的代号
3. 参数列表：使用该函数时，传入的参数类型。**形参**
4. 函数体语句
5. 函数返回值，`return`语句:和返回值类型相关，函数执行完后，返回相应的数据

**函数定义语法**

```
 返回值类型 函数名(参数列表){
 
    函数体语句;
    
    return表达式;
}
```

### 函数定义示例：加法实现及调用

````c++
int add(int num1, int num2){
    
    int sum = num1 + num2;
    
    return sum;
}
````

````c++
//函数的调用：函数名（实参）
#include<iostream>

using namespace std;

int a = 3;
int b = 4;

int result = add(a,b);

cout<<result;
````

## 函数中的值传递
* 所谓值传递，就是函数调用时实参将数值传入给形参
* 发生值传递时，如果形参发生改变，并不会影响实参的值

### 通过swap()函数来讨论值传递问题

````c++
//定义swap函数
void swap(int num1, int num2){//当不需要返回值时，不需要返回值类型。定义为`void`
    cout<<"交换前的形参数值:"<<endl;
    cout<<"num1 = "<<num1<<endl;
    cout<<"num2 = "<<num2<<endl;
    
    int temp = num1;
    
    num1 = num2;
    num2 = temp;
    
    cout<<"交换后的形参数值："<<endl;
    cout<<"num1 = "<<num1<<endl;
    cout<<"num2 = "<<num2<<endl;
}
````

````c++
//验证值传递
int numA = 10;
int numB = 20;

swap(numA, numB);
cout<<"交换后的实参值:"<<endl;
cout<<"numA = "<<numA<<endl;
cout<<"numB = "<<numB<<endl;
````

### 值传递问题解决：通过传入引用实现实参的改变


````c++
//定义swap函数
void swap(int *num1, int *num2){//当不需要返回值时，不需要返回值类型。定义为`void`
    cout<<"交换前的形参数值:"<<endl;
    cout<<"num1 = "<<*num1<<endl;
    cout<<"num2 = "<<*num2<<endl;
    
    int *temp = num1;
    
    num1 = num2;
    num2 = temp;
    
    cout<<"交换后的形参数值："<<endl;
    cout<<"num1 = "<<*num1<<endl;
    cout<<"num2 = "<<*num2<<endl;
}
````

## 函数的声明
**作用**：告诉编译器函数名称以及如何调用函数，函数的实际主体可以单独定义
>**函数的声明可以多次，但是函数的定义只能有一次**

````c++
int max(int num1, int num2){
    return num1 > num2 ? num1 :num2;
}
````

````c++
//函数声明的示例：比较函数，返回较大的数值
#include<iostream>

using namespace std;

cout<<"最大值为："<<max(20, 30);
````

## 函数的分文件编写
**作用**：让代码结构更加清晰

函数分文件编写的四个步骤：
1. 创建后缀名为.h的头文件
2. 创建后缀名为.cpp的源文件
3. 在头文件中写函数的声明
4. 在源文件中写函数的定义

### 函数分文件编写代码示例
```
//swap.h头文件
#include<iostream>

using namespace std;

void swap(int num1, int num2);

//swap.cpp源文件
#include"swap.h" //用户自定义头文件，用双引号

void swap(int num1, int num2){
    
    int temp = num1;
    num1 = num2;
    num2 = temp;
}
```
