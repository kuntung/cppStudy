# 数组的学习

## 数组定义及性质
1. 一维数组定义
    * 数据类型 数组名字[ ] = {数据1，数据2，数据3，...};
    * 数据类型 数组名字\[数组规模\] = {数据1，数据2，数据3，...};
    * 数据类型 数组名字[数组规模];
    
2. 数组性质
    * 数组占用一块连续的内存地址
    * 数组名字代表内存首地址

## 数组的应用及其实现
1. 冒泡排序：比较相邻的元素
    1. 如果第一个比第二个大，就交换这两个数
    2. 对每一对相邻元素做同样的工作，执行完毕后，找到第一个最大值
    3. 重复以上步骤，每次比较次数减1，直到不需要比较
    
    
>**算法分析**
>1. 算法复杂度：
>2. N个数，需要对比N-1轮，每轮对比次数从N-1逐步降低为1

```c++
//应用冒泡排序法对一个无序一维数组进行排序
#include<iostream>

using namespace std;

//int main(){}
int suffVar[] = {4, 2, 8, 0, 5, 7, 1, 3, 9};//初始化数组

cout<<"数组未进行排序前:"<<endl;
for(int i = 0; i<= sizeof(suffVar)/sizeof(int);i++){
    cout<<suffVar[i]<<" ";
}
cout<<endl;
//进行冒泡排序实现
int cnt = 0;
int currentMax = suffVar[0];
for(int j = sizeof(suffVar)/sizeof(int); j > 1;j--){
for(int i = 1; i< j;i++){
    if(suffVar[i-1] >= suffVar[i]){
        int temp = suffVar[i];
        suffVar[i] = suffVar[i-1];
        suffVar[i-1] = temp;
    }
    cnt++;
}
}
//黑马代码示例
// for(int i = 0; i < (sizeof(suffVar)/sizeof(int)-1);i++){
//     for(int j = 0; j < (sizeof(suffVar)/sizeof(int)-1-i);j++){
//         if(suffVar[j] >= suffVar[j+1]){
//             int temp = suffVar[j+1];
//             suffVar[j+1] = suffVar[j];
//             suffVar[j] = temp;
//     }
//         cnt++;
//     }
// }
    
    
cout<<"数组未进行排序后:"<<endl;
for(int i = 0; i< sizeof(suffVar)/sizeof(int);i++){
    cout<<suffVar[i]<<" ";
}
cout<<endl;
cout<<"共进行"<<cnt<<"次排序";
```

## 二维数组
1. 二维数组四种定义方式
    * 数据类型 数组名[行数][列数];
    * 数据类型 数组名[行数][列数] = {{数据1， 数据2}， {数据3，数据4}}； **最为直观的一种定义方式**
    * 数据类型 数组名[行数][列数] = {数据1，数据2，数据3，数据4}；
    * 数据类型 数组名[][列数] = {数据1， 数据2， 数据3， 数据4}；
    
2. 二维数组数组名
    * 查看二维数组所占内存空间，`sizeof()`
    * 获取二维数组首地址,`数组名`

```c++
//二维数组定义方式示例
int arr[2][3] = {{1,2,3},{4,5,6}}; //定义方式2

for(int i = 0;i< 2;i++){
    for(int j = 0;j<3;j++){
        cout<<arr[i][j]<<" ";
    }
    cout<<endl;
}
```

```c++
//二维数组定义方式示例2
int arr2[2][3] = {1, 2, 3, 4, 5, 6};

for(int i = 0;i< 2;i++){
    for(int j = 0;j<3;j++){
        cout<<arr2[i][j]<<" ";
    }
    cout<<endl;
}
```

```c++
//二维数组定义方式示例4
int arr3[][3] = {1, 2, 3, 4, 5, 6};

for(int i = 0;i< 2;i++){
    for(int j = 0;j<3;j++){
        cout<<arr3[i][j]<<" ";
    }
    cout<<endl;
}
```

### 二维数组数组名的用途

```c++
//代码示例
double arr[2][3] = {{1, 2, 3}, {4, 5, 6}};

cout<<"该二维数组所占内存空间大小为："<<sizeof(arr)<<endl;
cout<<"该二维数组第一行所占内存空间大小为："<<sizeof(arr[0])<<endl;
cout<<"该二维数组第一个元素所占内存空间大小为："<<sizeof(arr[0][0])<<endl;
cout<<"该二维数组元素个数为："<<sizeof(arr)/sizeof(arr[0][0])<<endl;
cout<<"该二维数组行数为："<<sizeof(arr)/sizeof(arr[0])<<endl;
cout<<"该二维数组列数为："<<sizeof(arr[0])/sizeof(arr[0][0])<<endl;
cout<<"该二维数组的首地址为:"<<arr<<endl;
cout<<"该二维数组的第一行首地址为:"<<arr[0]<<endl;
cout<<"该二维数组的第二行首地址为:"<<arr[1]<<endl;

// for(int i = 0; i<2; i++){
//     for(int j = 0; j < 3; j++){
//         cout<<arr[i][j]<<" ";
//     }
//     cout<<endl;
// }
```

### 二维数组应用案例
给定三个同学，分别输入并输出他们的语数外三科成绩。以表格的形式进行展示

```c++
//考试成绩录入及展示
#include<iostream>

using namespace std;

int score[3][3] = {{0, 0, 0}, {0, 0, 0}};

for(int i = 0; i < 3; i++){
    switch(i){
        case 0:
            cout<<"请输入张三的";
            break;
        case 1:
            cout<<"请输入李四的";
            break;
        case 2:
            cout<<"请输入王五的";
            break;
    }
    for(int j = 0; j < 3; j++){
        switch(j){
            case 0:
                cout<<"语文成绩：";
                break;
            case 1:
                cout<<"数学成绩：";
                break;
            case 2:
                cout<<"英语成绩：";
                break;
        }
        cin>>score[i][j];

    }
}
cout<<"他们的期末考试成绩如下表："<<endl;
cout<<"\t\t语文\t数学\t英语\t平均\t总和"<<endl;

for(int i = 0; i < 3; i++){
    int scoreSum = 0;
    if(0 == i){
        cout<<"\t张三";
    }
    else if(1 == i){
        cout<<"\t李四";
    }
    else{
        cout<<"\t王五";
    }
         for(int j = 0; j < 3; j++){
            cout<<"\t"<<score[i][j];
            scoreSum += score[i][j];
    }
        cout<<"\t"<<(float)scoreSum/3;
        cout<<"\t"<<scoreSum;
        cout<<endl;
    }
    
```
