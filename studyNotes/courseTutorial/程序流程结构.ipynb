## 三目运算符
作用：通过三目运算符实现简单的判断

语法：`表达式1？表达式2：表达式3`

解释：如果表达式1为真，则执行表达式2，并返回表达式2的结果

```c++
//三目运算符代码示例
#include<iostream>

using namespace std;
```

```c++
int num1 = 20;
int num2 = 10;
int num3 = 0;

num3 = (num1 > num2 ? num1 : num2);//输出返回num1和num2中数值较大者

cout<<num3;
```

## switch语句
语法表达式：

switch(表达式){

    case 结果1： 执行语句1； break；
    case 结果2： 执行语句2； break；
    ...
    default: 执行语句*； break；
    
}

---
>**`switch`**语句和**`if`**语句的区别：
>1. switch语句通常只能用于`整型`或者`字符型`变量，无法对`浮点型`变量进行判定
> 2. 但switch语句比if语句逻辑更加清晰明了

**给电影打分的程序实现**

根据用户输入的分值，判定用户对该电影的评价如何
1. 5以下，认为该片是个烂片
2. 5-7，一般
3. 7-9，还不错
4. 很好

```c++
#include<iostream>

using namespace std;
```

```c++
int score = 0;

cout<<"请输入对该片的评价分值";
cin>>score;
cout<<"您对该电影的评分为: "<<score<<endl;

switch(score){
    case 9:
        cout<<"您认为这部片子很不错！";
        break;
    case 7:
        cout<<"您觉得这部片子还可以！";
        break;
    default:
        cout<<"这是个垃圾电影！";
        break;
}
```

## 循环结构
1. while循环: while(循环条件){循环体}   
    * 注意循环条件不能常为真，否则会进入死循环
2. do-while循环:至少执行一次循环体（与while语句的不同）
3. for循环：for(起始表达式;条件表达式;末尾循环体){循环语句}
   * 针对给定的起始表达式
   * 判断条件表达式是否满足
   * 如果满足，则执行循环语句，再执行末尾循环体，最后再进行条件表达式判定
   * 如果不满足，则结束for循环语句
   

```c++
//while循环示例

int num = 0;
//输出0到10的整数
while(num <= 10){
    cout<<num++<<", ";
}
//用于测试前向递增和后向递增的差别
num = 0;
cout<<endl;
while(num <= 10){
    cout<<++num<<", ";
}
```

### 利用while循环设计一个猜数游戏
在每一次游戏开始时，电脑随机生成一个整数。并且根据用户输入的值判断是否猜对或者猜大了、猜小了

__随机数生成__
* rand()
* srand()//添加随机数种子，利用当前系统时间生成随机数，避免每次生成的随机数都一样
>srand((unsigned int)time(NULL)); //需要包含系统时间头文件<ctime>

```c++
//猜数游戏代码示例
#include<iostream>

using namespace std;

//系统生成一个随机数
int randNum = rand() % 100 + 1;//生成1-100之间的随机数
//提示用户输入一个猜测的值，并返回对比结果
int guessNum = -1;
int guessCnt = 0;

while(guessNum != randNum){
    cout<<"请输入一个1-100之间的整数值";
    cin>>guessNum;
    guessCnt++;
    if ((guessNum <1)||(guessNum > 100)){
        cout<<"猜的太离谱了！";
        cout<<endl;
    }
    else if(guessNum > randNum){
        cout<<"猜的结果过大！";
        cout<<endl;
    }
    else if(guessNum < randNum){
        cout<<"猜的结果过小！";
        cout<<endl;    
    }   
}
if(guessNum == randNum){
    cout<<"经过"<<guessCnt<<"次猜测，你终于猜对了。";
    cout<<endl;
    cout<<"正确答案为:"<<randNum<<endl;
}
```

### for循环语句示例：敲桌子
从1开始数到数字100，如果数字个位或者十位含有7，或者该数字是7的倍数，则打印字符串敲桌子。其余数字打印输出

__思路：__ 
1. 对于7的倍数，如果取模`%`结果为0，则表示该数是7的倍数
2. 对于十位或者个位有7，分别`/`,`%`10即可

```c++

#include<iostream>

using namespace std;

for(int num = 1; num<=100; num++){
    if((num/10) == 7 || (num%10)==7 ||(num/7*7) == num){
        cout<<"敲桌子"<<endl;
    }
    else{
        cout<<num<<endl;
    }
}
```
