# 结构体
结构体基本概念：结构体属于用户自定义的数据类型，允许用户存储不同的数据类型

## 结构体定义和使用
**语法** `struct 结构名{结构体成员列表};`

通过结构体创建变量的三种方式：
1. struct 结构体名 变量名
2. struct 结构体名 变量名 = {成员1值， 成员2值...}
3. 定义结构体时顺便创建变量

* 结构体定义的时候，关键字`struct`可以省略，结构体声明的时候，关键字可以省略
* 结构体变量通过`.`来访问变量属性

```c++
#include<iostream>
#include<string>

using namespace std;
```

```c++
//结构体定义示例
//创建学生数据类型：（姓名， 年龄， 分数）
//语法： struct 类型名称{成员列表 }
struct Student{
    
    string name;
    int age;
    float score;
}
```

```c++
//结构体声明type1
struct Student s1;//type1
//给s1属性赋值，通过`.`访问结构体变量中的属性
s1.age = 12;
s1.name = "唐昆";
s1.score = 100;
cout<<"这个学生的名字是："<<s1.name<<"\n"<<"年龄:"<<s1.age<<"\n"<<"分数："<<s1.score<<"\n";
```

```c++
//结构体声明type2
struct Student s2  = {"tangkun", 18, 99.9};//type2
cout<<"这个学生的名字是："<<s1.name<<"\n"<<"年龄:"<<s1.age<<"\n"<<"分数："<<s1.score<<"\n";
```

```c++
//结构体声明type3
struct Worker{string name; int age; float score;};
struct Worker w1;

w1.name = "张三";
w1.age = 18;
w1.score = 60.6;

// w2 = {"tangkun", 18, 99.9};

cout<<"这个工人的名字是："<<w1.name<<"\n"<<"年龄:"<<w1.age<<"\n"<<"分数："<<w1.score<<"\n";
```

## 结构体数组
**作用：**将自定义的结构体放入到数组中方便维护

**语法：**`struct 结构体名 数组名[元素个数] = {{struct1}, {}, ...{}};`

1、 先定义结构体

2、 再定义结构体数组

```c++
//结构体数组定义示例
//数据类型：struct Student
Student arr[3] = {
    {"张三", 18, 98},
    {"李四", 19, 95},
    {"王五", 20, 77}};
```

```c++
cout<<arr[0].name;
```

```c++
arr[0].name = "赵六";
cout<<arr[0].name;
```

## 结构体指针
**作用：**通过指针访问结构体中的成员
>通过操作符`->`可以通过结构体指针访问结构体属性

```c++
//结构体指针定义示例
//结构体定义
//创建学生数据类型：（姓名， 年龄， 分数）
//语法： struct 类型名称{成员列表 }
struct Student{
    
    string name;
    int age;
    float score;
}
```

```c++
//创建学会结构体变量
struct Student s1 = {"张三", 18, 100};

//结构体变量指针
struct Student *ps1 = &s1; //通过指针指向结构体变量
```

```c++
cout<<ps1->name;
cout<<"\n";
cout<<s1.age;
```

## 结构体嵌套结构体
即结构体中的成员可以是另一个结构体

```c++
//结构体嵌套结构体示例
//定义一个学校结构体
struct School{
    int level;//985, 211
    string name;
    string location;
};

//定义一个研究生结构体
struct Yanjiusheng{
    long id;
    string name;
    string gender;
    int age;
    School schName;
};

//定义一个教师结构体
struct Teacher{
    int id;
    string name;
    int age;
    School schName;
    Yanjiusheng stuName;
};
```

```c++
cout<<tangkun->name<<"是"<<tangkun->schName.name<<"的带学生。"<<"\n";
cout<<"这是一所"<<uestc.level<<"带学。"<<"位于"<<tangkun->schName.location;
cout<<"\n";
cout<<"他的学号是:"<<tangkun->id<<"\n";
cout<<"他今年"<<tangkun->age<<"岁\n";
cout<<"他是个帅气的"<<tangkun->gender<<"孩子.\n";
```

## 结构体做为函数参数
**作用：**将结构体作为参数向函数中传递

传递方式有两种：
* 值传递
* 地址传递

`值传递`和`地址传递`的区别：
1. 通过值传递无法改变原结构体的属性
2. 通过地址传递可以改变原结构体的属性

```c++
#include<iostream>
#include<string>

using namespace std;
```

```c++
//结构体作为函数参数代码示例
//将学生结构体传入到一个参数中，打印学生身上的所有信息

//定义学生结构体
struct StudentPar{
    string name;
    int age;
    float score;
};
```

```c++
//打印学生信息的函数
//值传递
void printStudent1(struct StudentPar s){
    cout<<s.name<<"的年龄为:"<<s.age<<"分数为:"<<s.score<<"\n";
    s.name = "李四";
}
```

```c++
//地址传递
void printStudent2(struct StudentPar *s){
    cout<<s->name<<"的年龄为:"<<s->age<<"分数为:"<<s->score<<"\n";
    s->name = "李四";
}
```

```c++
struct StudentPar stu1;

stu1.name = "张三";
stu1.age = 19;
stu1.score = 79.8;

StudentPar *stuP = &stu1;

printStudent1(stu1);//将结构体作为参数传入并打印显示
cout<<stu1.name<<"的年龄为:"<<stu1.age<<"分数为:"<<stu1.score<<"\n";
printStudent2(&stu1);
cout<<stuP->name<<"的年龄为:"<<stu1.age<<"分数为:"<<stuP->score<<"\n";//通过地址传递成功实现了结构体属性变量的修改
```

## 结构体中const使用场景
**作用：**用`const`来防止误操作

```c++
//地址传递，指针只占用四个或者八个字节。减少内存占用空间。
//而且不会生成新的副本
void printStudent2(const struct StudentPar *s){
    cout<<s->name<<"的年龄为:"<<s->age<<"分数为:"<<s->score<<"\n";
//     s->name = "李四";
}
//加上const，一旦执行修改操作，就会报错。
//好处：可以防止误操作
```

```c++
struct StudentPar stu = {"张三", 18, 97.8};

printStudent2(&stu);
```
